```python
import pandas as pd
import numpy as np
```

Download and Clean Dataset

```python
concrete_data = pd.read_csv('https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DL0101EN/labs/data/concrete_data.csv')
concrete_data.head()
```

Seperate the concrete_data's "gonna use in train" and "gonna use in prediction" parts ğŸ™‚

```python
concrete_data_columns = concrete_data.columns

predictors = concrete_data[concrete_data_columns[concrete_data_columns != 'Strength']]
target = concrete_data['Strength']
```

First 5 of predictors and target

```python
predictors.head()
target.head()
n_cols = predictors.shape[1]
```

Normalize the data

```python
predictors_norm = (predictors - predictors.mean()) / predictors.std()
target_norm = (target - target.mean()) / target.std()
n_cols_norm = predictors_norm.shape[1]
```

Import KERAS

```python
from keras.models import Sequential
from keras.layers import Dense
```

Model Functions

```python
# define regression model
def regression_model_normalized(mark):
    # create model
    model = Sequential()
    if mark!="d":
        model.add(Dense(10, activation='relu', input_shape=(n_cols_norm,)))
    else:
        model.add(Dense(10, activation='relu', input_shape=(n_cols_norm,)))
        model.add(Dense(10, activation='relu'))
        model.add(Dense(10, activation='relu'))
    model.add(Dense(1))
    
    # compile model
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model

def regression_model_non_normalized():
    # create model
    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=(n_cols_norm,)))
    model.add(Dense(1))
    
    # compile model
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
```

Define MSE Lists

```python
l_mean_squared_error = [] # LÄ°ST OF MSE WÄ°TH NON-NORMALÄ°ZED DATA
l_mean_squared_error_N =[] # LÄ°ST OF MSE WÄ°TH NORMALÄ°ZED DATA
l_mean_squared_error_N_100 =[] # LÄ°ST OF MSE WÄ°TH NORMALÄ°ZED DATA & TRAÄ°NÄ°NG EPOCH 100 INSTEAD OF 50
l_mean_squared_error_N_marked =[] # LÄ°ST OF MSE WÄ°TH NORMALÄ°ZED DATA & THERE IS 3 HIDDEN LAYERS INSTEAD OF 1 HIDDEN LAYER
```

Import SCÄ°KÄ°T-LEARN

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

Train & Evaulate Model (Normalized & Non-Normalized)

```python
def normalized(mark="x",e=50):
    for i in range(50):
        predictor_X , predictor_Y = train_test_split(predictors_norm,test_size=.3,random_state=42)
        target_X, target_Y = train_test_split(target_norm,test_size=.3,random_state=42)

        model = regression_model_normalized(mark=mark)

        model.fit(predictor_X, target_X, epochs=e, verbose=0)

        y = model.predict(predictor_Y)

        mse = mean_squared_error(target_Y, y)
        if mark!="d":
            if e!=100:
                l_mean_squared_error_N.append(mse)
            else:
                l_mean_squared_error_N_100.append(mse)
        else:
            l_mean_squared_error_N_marked.append(mse)

def non_normalized():
    for i in range(50):
        predictor_X , predictor_Y = train_test_split(predictors,test_size=.3,random_state=42)
        target_X, target_Y = train_test_split(target,test_size=.3,random_state=42)

        model = regression_model_non_normalized()

        model.fit(predictor_X, target_X, validation_split=0.3, epochs=50, verbose=0)

        y = model.predict(predictor_Y)

        mse = mean_squared_error(target_Y, y)
        l_mean_squared_error.append(mse)
```

Printing =><br>
List of Mean Squared Error (as lmse)<br>
Mean of "lsme"<br>
Standart Devitaion of "lsme"<br>

```python
non_normalized()
MSEs = pd.DataFrame(l_mean_squared_error)
mean_MSEs= MSEs.mean()
std_MSEs = MSEs.std()
```

```python
normalized()
MSEs_N = pd.DataFrame(l_mean_squared_error_N)
mean_MSEs_N= MSEs_N.mean()
std_MSEs_N = MSEs_N.std()
```

```python
normalized(e=100)
MSEs_N_100 = pd.DataFrame(l_mean_squared_error_N_100)
mean_MSEs_N_100= MSEs_N_100.mean()
std_MSEs_N_100 = MSEs_N_100.std()
```

```python
normalized(mark="d")
MSEs_N_marked = pd.DataFrame(l_mean_squared_error_N_marked)
mean_MSEs_N_marked= MSEs_N_marked.mean()
std_MSEs_N_marked = MSEs_N_marked.std()
```

Comparing The (MSEs, mean_MSEs, std_MSEs) Of Normalized And Non-normalized Data
